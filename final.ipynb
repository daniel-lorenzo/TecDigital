## Ejercicio 1
Calcular las secuencias de 16 bits dadas su representación hexadecimal.

    A = 0000 0110 0111 0000
    B = 0110 0100 1101 0010
    C = 1110 0101 1101 1011

### Solución

    A = 0000 0110 0111 0000        bin
           0    6    7    0        hexa
    B = 0110 0100 1101 0010        bin
           6    4    D    2        hexa
    C = 1110 0101 1101 1011        bin
           E    5    D    B        hexa

```python
# Python code
# Converte binario a decimal
A = int('0000011001110000', 2)
B = int('0110010011010010', 2)
C = int('1110010111011011', 2)
# Convierte decimal a hexadecimal            
A_hex = hex(A)
B_hex = hex(B)
C_hex = hex(C)
# Convierte decimal a octal
A_oct = oct(A)
B_oct = oct(B)
C_oct = oct(C)
# Imprime resultados
print('     Hexa  | Decimal | Octal')
print('A =  %s |  %s   |   %s'%(A_hex,A,A_oct) )
print('B = %s | %s   |  %s'%(B_hex,B,B_oct) )
print('C = %s | %s   | %s'%(C_hex,C,C_oct) )
```

## Ejercicio 2
Indicar el rango de un número de 8 bits según las codificaciones siguientes.
1. Binario puro
2. Complemento a 2

### Solución

El rango de magnitud de un número binario depende del número de bits (n). Utilizando un byte u ocho bits se pueden representar 256 números diferentes. Con 16 bits se pueden representar 65536 números diferentes.
La fórmula para calcular el número de combinaciones diferentes de n bits es:
Número total de combinaciones = $2^n$

Para los números con signo en complemento a 2, el rango de valores para números de n bits es:
Rango = $-\left( 2^{n-1} \right)$ hasta $+ \left(2^{n-1}-1 \right)$


(1) __Binario puro:__  

$2^n \Rightarrow  2^8=256$

__Rango:__ 0 hasta 256

(2) __Complemento a 2__

$\begin{array}{l}
- \left( 2^{n-1} \right) = - \left( 2^{8-1} \right) = -128 \\
+ \left( 2^{n-1}-1 \right) =+ \left( 2^{8-1}-1 \right) = 127 
\end{array}$

__Rango:__ -128 hasta +127

Como vamos a usar un flip-flop tipo J-K entonces usamos su tabla de verdad.

$$ \begin{array}{|cc|cc|}

1. Que se el estado actual es 0, el estado futuro se tomará 1 cuando $J$ sea 1.

3. En cualquier otra condición es estado actual se conserva en el futuro.

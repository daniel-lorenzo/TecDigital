## Ejercicio 1
Calcular las secuencias de 16 bits dadas su representación hexadecimal y octa.

    A = 0000 0110 0000 0111
    B = 0000 0000 1101 0110
    C = 1001 0000 0000 1010

### Solución
Para pasar a hexa tomamos grupos de cuatro y para pasar a octa tomamos grupos de tres.
## $$ A = \underbrace{ \overbrace{0000}^\mathrm{nibble} \; 0110 \; \overbrace{0000 \; 0111}^\mathrm{byte} }_\mathrm{word} $$

    A = 0000 0110 0000 0111 bin    |   A = 000 011 000 000 111 bin
          0    6    0    7  hexa   |        0   3   0   0   7  octa
    
    B = 0000 0000 1101 0110 bin    |   B = 000 000 011 010 110 bin
          0    0   D     6  hexa   |        0   0   3   2   6  octa    
    
    C = 1001 0000 0000 1010 bin    |   C = 001 001 000 000 001 010 bin
          9    0    0    A  hexa   |        1   1   0   0   1   2  octa

```python
# Python code
# Convierte binario a decimal
A = int('0000011000000111',2)
B = int('0000000011010110',2)
C = int('1001000000001010',2)
# Convierte decimal a hexadecimal
A_hex = hex(A)
B_hex = hex(B)
C_hex = hex(C)
# Convierte a octal
A_oct = oct(A)
B_oct = oct(B)
C_oct = oct(C)
# Imprime resultados
print('     Hexa  |  Decimal  |  Octal')
print('A = %s  |  %s     |  %s'%(A_hex,A,A_oct))
print('B = %s   |  %s      |  %s'%(B_hex,B,B_oct))
print('C = %s |  %s    |  %s'%(C_hex,C,C_oct))
```

## Ejercicio 2
Pasar a BCD

    A = 0110 1001 0111 1000
    B = 0000 0010 0101 0010 0110

### Solución
Pasamos A a decimal
$$ A = 2^3 + 2^4 + 2^5 + 2^6 + 2^8 + 2^{11} + 2^{13} + 2^{14} $$

```python
A = 2**3 + 2**4 + 2**5 + 2**6 + 2**8 + 2**11 + 2**13 + 2**14
print('A = %s'%A)
```

```python
A = int('0110100101111000',2)
print('A = %s'%A)
```

    A = 0110 1001 0111 1000  (binario)
    
    A = 0010 0111 0000 0000  (BCD)
          2    7    0    0   (Decimal)

Pasamos B a decimal
$$ B = 2^1 + 2^2 + 2^5 + 2^8 + 2^{10} + 2^{13} $$

```python
B = 2**1 + 2**2 + 2**5 + 2**8 + 2**10 + 2**13
print('B = %s'%B)
```

```python
B = int('0010010100100110',2)
print('B = %s'%B)
```

    B = 0000 0010 0101 0010 0110  (binario)
    
    B = 1001 0101 0001 0000       (BCD)
          9    5    1    0        (Decimal)
